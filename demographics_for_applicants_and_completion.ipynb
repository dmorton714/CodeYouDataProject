```python
employed_demo_df = pd.merge(demo_and_program_df, employment_df, on="Auto ID", how="inner")
full_df = pd.merge(employed_demo_df, applicants_df, on="Auto ID", how="inner")

full_df
```

##### Weeding out similar rows with a priority list

```python
# heirarchy of what we most want, 0 is what we want
priority = {
    "Successfully Completed": 0,
    "Did Not Complete": 1,
    "Not Provided": 2
}

# creating a new column to map on the ranks of 'priority'
full_df['outcome_rank'] = full_df["Outcome"].map(priority)

# sorting the priorties from 0 to 2
df_sorted = full_df.sort_values(["Auto ID", "outcome_rank"])

# removing whatever is listed first
best_per_id = df_sorted.drop_duplicates(subset=["Auto ID"], keep='first')

# dropping the extra column we used
best_per_id = best_per_id.drop(columns="outcome_rank")

best_per_id
```

##### Organizing

```python
# drop Race_y, Gender_y, Veteran_y
dropped_df = best_per_id.drop(columns=["Race_y", "Gender_y", "Veteran_y"])

# rename the remainders
fixed_df = dropped_df.rename(columns={
    "Race_x": "Race",
    "Gender_x": "Gender",
    "Veteran_x": "Veteran",
    }
)
```

##### Fix 'Birthdate' to show age

```python
# creating the function to figure out their age
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

fixed_df["Birthdate"] = fixed_df["Birthdate"].apply(age)

fixed_df.rename(columns={
    "Birthdate": "Age"
})

display(fixed_df)
```

##### Adding enrollments

Self-explanatory.

```python
fullest_df = pd.merge(fixed_df, enrollment_df, on="Auto ID", how="inner")

fullest_df
```
