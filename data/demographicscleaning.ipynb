# Cleaning and Exploring the Demographics and Programs data

```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
df = pd.read_excel('All demographics and programs.xlsx')
```

```python
# drop the first name and last name columns since they don't have any names
# drop single parent because it is empty (only 1 row has 'No')
# drop Ethnicity Hispanic/Latino because it is all null
# drop Ex-Offender because only 1 row has No and the rest are null
# combined Transgender M to F and F to M to the category 'Transgender' because there were only 
# 16 that identifed M to F or F to M


def data_cleaning(df_clean):
    df_clean = df.drop(columns=['First Name','Last Name','Ethnicity Hispanic/Latino',
                     'Single Parent','Ex-Offender','Program: Program Name'])
    df_clean = df_clean['Gender'].str.replace('Transgender female to male','Transgender')
    df_clean = df_clean['Gender'].str.replace('Transgender male to female','Transgender')
    splitting = df_clean['Race'].str.split(';', expand=True)
    splitting.columns = [f'Race_{i+1}' for i in range(splitting.shape[1])]
    df_clean = pd.concat([df_clean, splitting], axis=1)
    df_clean.drop(columns=['Race'])
    return(df_clean)

df_clean
```
